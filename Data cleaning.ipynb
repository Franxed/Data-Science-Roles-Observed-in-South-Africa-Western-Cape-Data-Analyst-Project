# Cleaning Locations.

### I notice there is location inconsistencies. For e.g. 3 different ways of telling it's in 'Cape Town'.

```python
df.location.value_counts()
```

### First, I'm splitting and selecting the first values as this is the only data we need. (Everything is based already in WC/Western Cape ZA)

```python
df['Location'] = df['location'].apply(lambda x: x.split(',')[0])
df['Location'].value_counts()
```

### Some locations are suburbs of Cape Town.

```python
suburbs_ct = {'Bellville': 'Cape Town', 'Somerset West': 'Cape Town', 'Brackenfell': 'Cape Town',
    'Milnerton': 'Cape Town', 'Century City': 'Cape Town', 'Durbanville': 'Cape Town',
    'Pinelands': 'Cape Town', 'Claremont': 'Cape Town', 'Rosebank': 'Cape Town',
    'Nuweland': 'Cape Town', 'Wynberg': 'Cape Town', 'Southern Suburbs': 'Cape Town',
    'Goodwood': 'Cape Town', 'Sea Point': 'Cape Town', 'Blackheath': 'Cape Town',
    'Rondebosch': 'Cape Town', 'Kenilworth': 'Cape Town', 'Northern Suburbs': 'Cape Town',
    'Tygervalley': 'Cape Town', 'City of Cape Town': 'Cape Town'}
df['City'] = df['Location'].replace(suburbs_ct).sort_index()
df.City.value_counts()
```

```python
pd.set_option('display.max_rows', None) # I want to view the FULL dataframe.
df
```

### After confirming the locations, I drop the unformatted 'location' column.

```python
df = df.drop(columns='location')
df
```

# Assigning Seniority of Job Positions to Level Column.

```python
def get_level(title):
    title = str(title).lower()
    if 'senior' in title or 'snr' in title:
        return 'Senior'
    elif 'junior' in title or 'jnr' in title or 'graduate' in title:
        return 'Junior'
    elif 'associate' in title:
        return 'Associate/Entry'
    else:
        return 'Mid-Level'

df['Level'] = df['title'].apply(get_level)
```

```python
df['Level'].value_counts()
```

# Formatting Job Type Column.

## Seperating job_type by Full-Time, Part-Time & Internship.

```python
df.job_type.value_counts()
```

### First in removing the confusion 'parttime, fulltime' by converting them to full-time.

```python
df.job_type = df.job_type.replace({"parttime, fulltime":"fulltime"})
df.job_type.value_counts()
```

### Then I format the strings/title to make it easier to read.

```python
def get_job_type(job_type):
    job_type = str(job_type).lower().strip()
    if 'fulltime' in job_type:
        return "Full-Time"
    elif 'parttime' in job_type:
        return "Part-Time"
    elif 'internship' in job_type:
        return "Internship"
    elif 'unknown' in job_type:
        return "Unknown"
df['Job Type'] = df['job_type'].apply(get_job_type)
df['Job Type'].value_counts()
```

```python
df
```

### Now I'll drop job_type column as it's no longer needed.

```python
df = df.drop(columns='job_type')
df
```

# Cleaning Description Column.

### Next I want clean the description column by removing characters and strange syntax.

```python
df['description']
```

### Using Python Regex for specifying patterns.

```python
patterns = r'\\n|\n|\*{1,3}|=+|#{1,6}\s*|---+|/'

df['Description'] = (
    df['description']
    .str.replace(patterns, ' ', regex=True)
    .str.replace(r'\\-', '-', regex=True)
    .str.replace(r'\\&', '&', regex=True)    # \& → &
    .str.replace(r'\\|', '', regex=True)     # \| → remove
    .str.replace(r'\s{2,}', ' ', regex=True)
    .str.strip()
    .replace('Unknown', None)
)
```

### Seems like every weird characters are removed ;)

```python
print(df['Description'].iloc[0])
```

### Dropping the old description column.

```python
df = df.drop(columns='description')
df
```
